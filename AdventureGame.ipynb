```python
class character:
    def __init__(self,health,sheild,damage,name,money,Type):
        self.health=health
        self.sheild=sheild
        self.damage=damage
        self.name=name
        self.money=money
        self.died=False
        self.Type=Type

    def damageTaken(self,damageTaken):
        if self.sheild>0 and self.sheild==True:
            self.sheild-=damageTaken
        elif self.sheild<=0 or self.sheild==False:
            self.sheild=False
            self.health-=damageTaken
            if self.health<=0:
                self.died=True
        
    def died(self):
        while self.died==True:
            print(f'\nYOU DIED\nCONTINUE FOR 25 CENTS?')
            choice=input('yes or no')
            if choice=='yes':
                if self.money>=25:
                    print('more playing')
                    self.died=False
                elif self.money<25:
                    print('YOU DO NOT HAVE ENOUGH MONEY')
            elif choice=='no':
                print('THANKS FOR PLAYING')
                pass
            else:
                pass
```
